# Pyplot Tutorial:

matplotlib.pyplot is a collection of command style function that make matplotlib work like MATLAB.

Each pyplot function makes some change to a figure: e.g, creates a figure, creates a plotting area in a figure, plots some linesin a plotting area, decorates the plot with labels,etc.

```python
import matplotlib.pyplot as plt
```

```python
print(plt.plot([2,4,6,4])) #y-axis
plt.ylabel('Numbers')
plt.xlabel('Indices') #index denotes the x-axis
plt.title('MyPlot')

plt.show
```

If you provide a single list or array to the plot () command , matplotlib assumes it is a sequence of y values, and automatically generates the values for you. Since python ranges start with 0 , the default x vector has the same length as y but starts with 0 , Hense the data x are [0,1,2,3].


PLOT x versus Y

```python
plt.plot([-4,-2,0,2,4],[16,4,0,4,16]) #here x-axis is [1,2,3,4] and y axis is [1,4,9,16]
plt.ylabel('sqaures')
plt.xlabel('numbers')
plt.grid() #grid on

plt.show()
```

For every x,y pairs of arguments , there is an optional third argument which is the
format string that indicates the oclor and line type of plot.

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'ro') #'r - means red color and o - non-connecting dots'
plt.grid()
plt.xlabel('numbers')
plt.ylabel('squares')
plt.show()
```

if matplotlib were limited to working with lists, it would be fairly uselss for numeric processing.Generally , you will use numpy arrays. In fact, all sequences are converted to numpy arrays internally.

```python
import numpy as np
```

```python
t = np.arange(0. , 5. , 0.2)
print(t)
print('\n')
#blue dashes , red squares and greeen triangles
plt.plot(t,t**2, 'b--' , label = '^2') #b-- indicates blue color and dotted line
plt.plot(t,t**2.2 , 'rs', label = '^2.2') #rs -red color and square in shape
plt.plot(t, t**2.5 , 'g^' , label = '^2.5') #g - green and ^ symbolise triangle.
plt.grid()
plt.legend() # add legend based on line labels[left-top]
plt.show()
```

## Controlling line properties 

```python
"""USE KEYWORD ARGS"""
```

```python
x = [2,3,4,5]
y = [4,9,16,25]

plt.plot(x,y,linewidth = 5.0) #draw width of line
plt.show()
```

```python
use the setp()
```

```python
num = [1,2,3,4]
sqr = [1,4,9,16]
num2 = [1,2,3,4]
mul = [2,4,6,8]

lines = plt.plot(num,sqr,num2,mul)

#use keyword args
#Lines[0] :refers to the num,sqr
#lines[1]: refers to the num2, mul
plt.setp(lines[0], color ='r' , linewidth = 2.0)#setp -set properties ..  

#or MATLAB style string value paris
plt.setp(lines[1], 'color', 'g' , 'linewidth' ,2.0)# color = 'r' is similar to 'color', 'g'

plt.grid()
```

```python
"""working with multiple figures and axes"""
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0 , 5.0 , 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
#The subplot command specifies num rows , numcolumns,
#fignum where fignum ranges form 1 to numrows * numcols.
plt.subplot(211) #2 - no of rows , 1 - no of columns , 1 - figure 1
plt.grid()
plt.plot(t1,f(t1),'b--') #bule and dashed line


plt.subplot(212)  
plt.plot(t2,np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
plt.figure(1)                  # space for the first figure
plt.subplot(211)               # the first subplot in the first figure
plt.plot([1,2,3])              
plt.subplot(212)               # the second object in the first figure
plt.plot([4,5,6])

plt.figure(2)                  # a second figure
plt.plot([4,5,6])              # creates a subplot(111) by default

plt.figure(1)                  # figure 1 current; subplot(212) still current
plt.subplot(211)               # make subplot(211) in figure` current
plt.title('matplotlib is easse')
plt.show()

```
