### Packages

packages are a way of structing Python's module namespace by using 
"dotted module names".

A directory must contain a file named init.py in order for Python to consider it as a package. This file can be left empty but we generally place the initialization code for that package in this file.



##### IMPORTING MODULE FROM PACKAGES


We can import modules from package using dot(.) operators

```python
#import game.Image.open

#game is package (folder)
#image is sub-package (sub-folder)

#every package and sub-package contains __init__.py 
#open is module (.py file)
```
